```python
import pandas as pd
import numpy as np
```

```python
cdf = pd.read_csv("classification.csv")
```

```python
sdf = pd.read_csv("Stock Analysis Report - Data Source (1).csv")
```

```python
columns_to_convert = [
    '0 - 7 Days Avg', '8 - 15 Days Avg', '16 - 30 Days Avg', 
    '31 - 45 Days Avg', '46 - 60 Days Avg', '61 - 90 Days Avg', 
    '91 - 120 Days Avg', '121 - 150 Days Avg', '151 - 180 Days Avg', 
    'Older Than 180 Days Avg', 'Current PO Price'
]

sdf[columns_to_convert] = sdf[columns_to_convert].replace({',': ''}, regex=True).astype('int64')

    
sdf.info()
```

```python
Formulated_df = sdf[['PID', 'Product Name', 'Product Age', 'Sub Category']].copy()
Formulated_df
```

```python
ranks = []
Formulated_df = sdf.merge(cdf , how = 'left', on = ['PID'])
for i in range(len(Formulated_df)):
    if Formulated_df['Sub Category'][i] == 'Games':
        ranks.append(Formulated_df['Rank'][i])
    if Formulated_df['Sub Category'][i] == 'Accessories':
        ranks.append(11)
    elif Formulated_df['Sub Category'][i] == 'Consoles':
        ranks.append(12)
    elif Formulated_df['Sub Category'][i] == 'PCComponents':
        ranks.append(13) 
Formulated_df['Rank'] = ranks        
Formulated_df
```

```python
 Formulated_df[['0 - 7 Days Count','8 - 15 Days Count', '16 - 30 Days Count', '31 - 45 Days Count',
       '46 - 60 Days Count', '61 - 90 Days Count', '91 - 120 Days Count',
       '121 - 150 Days Count', '151 - 180 Days Count',
       'Older Than 180 Days Count', '0 - 7 Days Avg', '8 - 15 Days Avg',
       '16 - 30 Days Avg', '31 - 45 Days Avg', '46 - 60 Days Avg',
       '61 - 90 Days Avg', '91 - 120 Days Avg', '121 - 150 Days Avg',
       '151 - 180 Days Avg', 'Older Than 180 Days Avg']]
```

```python
cs_name = []
iwl = []
for i in range(len(Formulated_df)):
    if Formulated_df['Rank'][i] == 1:
        cs_name.append('Volatile Games')
    elif Formulated_df['Rank'][i] == 2:
        cs_name.append('All time Hits')
    elif Formulated_df['Rank'][i] == 3:
        cs_name.append('Moderate')
    elif Formulated_df['Rank'][i] == 4:
        cs_name.append('Low')
    elif Formulated_df['Rank'][i] == 5:
        cs_name.append('NS Games')
    elif Formulated_df['Rank'][i] == 6:
        cs_name.append('NS Junk')
    elif Formulated_df['Rank'][i] == 7:
        cs_name.append('Junk A')                      
    elif Formulated_df['Rank'][i] == 8:
        cs_name.append('Junk B')
    elif Formulated_df['Rank'][i] == 9:
        cs_name.append('Junk C')
    elif Formulated_df['Rank'][i] == 10:
        cs_name.append('Junk D')   
    elif Formulated_df['Rank'][i] == 11:
        cs_name.append('Accessories')   
    elif Formulated_df['Rank'][i] == 12:
        cs_name.append('Consoles')   
    elif Formulated_df['Rank'][i] == 13:
        cs_name.append('PCComponents')
    else:
        cs_name.append('NaN')
Formulated_df['Class Name'] = cs_name     
```

```python
iwl = []
vwl = []
iol = []
vol = []
i90 = []
iv90 = []
i_45_90 = []
iv_45_90 = []
i_45 = []
for i in range(len(Formulated_df)):
    rank_value = Formulated_df.at[i, 'Rank']
    
    if rank_value == 1:
        
        
        
        iwl_sum = Formulated_df.at[i, '0 - 7 Days Count'] + Formulated_df.at[i, '8 - 15 Days Count']
        iwl.append(iwl_sum)
        
        vwl_sumproduct = (Formulated_df.at[i, '0 - 7 Days Count'] * Formulated_df.at[i, '0 - 7 Days Avg']) + \
                  (Formulated_df.at[i, '8 - 15 Days Count'] * Formulated_df.at[i, '8 - 15 Days Avg'])
        vwl.append(vwl_sumproduct)
        
        iol_sum = Formulated_df.iloc[i, 9:17].sum()  
        iol.append(iol_sum)
        
        count_values = Formulated_df.iloc[i, 9:17].values
        avg_values = Formulated_df.iloc[i, 19:27].values
        vol_sumproduct = np.sum(np.multiply(count_values,avg_values))
        vol.append(vol_sumproduct)

        
        
        
    elif 5 <= rank_value <= 7:
        iwl_sum = Formulated_df.at[i, '0 - 7 Days Count'] + Formulated_df.at[i, '8 - 15 Days Count'] + \
                  Formulated_df.at[i, '16 - 30 Days Count'] + Formulated_df.at[i, '31 - 45 Days Count']
        iwl.append(iwl_sum)
        
        vwl_sumproduct=  (Formulated_df.at[i, '0 - 7 Days Count'] * Formulated_df.at[i, '0 - 7 Days Avg']) + \
                  (Formulated_df.at[i, '8 - 15 Days Count'] * Formulated_df.at[i, '8 - 15 Days Avg']) + \
                  (Formulated_df.at[i, '16 - 30 Days Count'] * Formulated_df.at[i, '16 - 30 Days Avg']) + \
                  (Formulated_df.at[i, '31 - 45 Days Count'] * Formulated_df.at[i, '31 - 45 Days Avg'])
        vwl.append(vwl_sumproduct)
        
        iol_sum = Formulated_df.iloc[i, 11:17].sum() 
        iol.append(iol_sum)
        
        count_values = Formulated_df.iloc[i, 11:17].values
        avg_values = Formulated_df.iloc[i, 21:27].values
        vol_sumproduct = np.sum(np.multiply(count_values,avg_values))
        vol.append(vol_sumproduct)
        
        
    elif 8 <= rank_value <= 10:
        iwl_sum = 0
        iwl.append(iwl_sum)
        
        vwl_sumproduct = 0
        vwl.append(vwl_sumproduct)
        
        iol_sum = Formulated_df.iloc[i, 7:17].sum()
        iol.append(iol_sum)
        
        count_values = Formulated_df.iloc[i, 7:17].values
        avg_values = Formulated_df.iloc[i, 17:27].values
        vol_sumproduct = np.sum(np.multiply(count_values,avg_values))
        vol.append(vol_sumproduct)
        
        
        
    elif 2 <= rank_value <= 13:
        iwl_sum = Formulated_df.at[i, '0 - 7 Days Count'] + Formulated_df.at[i, '8 - 15 Days Count'] + \
                  Formulated_df.at[i, '16 - 30 Days Count']
        iwl.append(iwl_sum)
        
        vwl_sumproduct = (Formulated_df.at[i, '0 - 7 Days Count'] * Formulated_df.at[i, '0 - 7 Days Avg']) + \
                  (Formulated_df.at[i, '8 - 15 Days Count'] * Formulated_df.at[i, '8 - 15 Days Avg']) + \
                  (Formulated_df.at[i, '16 - 30 Days Count'] * Formulated_df.at[i, '16 - 30 Days Avg'])
        vwl.append(vwl_sumproduct)
        
        iol_sum = Formulated_df.iloc[i, 10:17].sum()
        iol.append(iol_sum)
        
        count_values = Formulated_df.iloc[i, 10:17].values
        avg_values = Formulated_df.iloc[i, 20:27].values
        vol_sumproduct = np.sum(count_values * avg_values)
        vol.append(vol_sumproduct)
        
    else:
        iwl_sum = 'NA'
        iwl.append(iwl_sum)
        vwl_sumproduct = 'NA'
        vwl.append(vwl_sumproduct)
        vol_sumproduct = 'NA'
        vol.append(vol_sumproduct)
        
    inv_90 = Formulated_df.iloc[i,13:17].sum()
    i90.append(inv_90)
    
    
    count_values_90 = Formulated_df.iloc[i, 13:17].values
    avg_values_90 = Formulated_df.iloc[i, 23:27].values
    iv90_sumproduct = np.sum(np.multiply(count_values_90 , avg_values_90))
    iv90.append(iv90_sumproduct)
    
    inv_45_90 = Formulated_df.iloc[i, 11:13].sum()
    i_45_90.append(inv_45_90)
    
    count_values_45_90 = Formulated_df.iloc[i, 11:13].values
    avg_values_45_90 = Formulated_df.iloc[i, 21:23].values
    iv_45_90_sumproduct = np.sum(np.multiply(count_values_45_90, avg_values_45_90))
    iv_45_90.append(iv_45_90_sumproduct)
    
    
    i_45_sum = Formulated_df.iloc[i, 11:17].sum() 
    i_45.append(i_45_sum)


Formulated_df['Inv Within Limit'] = iwl

Formulated_df['Value Within Limit'] = vwl

Formulated_df['Value Outside Limit'] = vol

Formulated_df['Inv Outside Limit'] = iol
  
Formulated_df['Total Purchase Value'] =  Formulated_df['Total Qty'] * Formulated_df['Avg Inventory Value']

Formulated_df['Expected Revenue'] = Formulated_df['Total Qty'] * Formulated_df['Current PO Price']

Formulated_df['Margin'] = Formulated_df['Expected Revenue'] - Formulated_df['Total Purchase Value']

Formulated_df['Margin %'] = (Formulated_df['Margin'] / Formulated_df['Expected Revenue']).round(2) 

Formulated_df['Inv > 90 Days'] = i90

Formulated_df['Inv Value >90'] = iv90

Formulated_df['Inv 45 - 90 Days'] = i_45_90

Formulated_df['Inv Value 45-90'] = iv_45_90

Formulated_df['Inv >45'] = i_45

Formulated_df[['PID','Product Name', 'Product Age', 'Category_x', 'Sub Category', 'Rank', 'Class Name', 'Inv Within Limit',
               'Value Within Limit', 'Inv Outside Limit', 'Value Outside Limit', 'Total Purchase Value', 'Expected Revenue',
               'Margin','Margin %', 'Inv > 90 Days','Inv Value >90', 'Total Qty', 'Inv 45 - 90 Days', 'Inv Value 45-90', 
               'Inv >45']]
```

```python
inventory_df = sdf.iloc[:,7:27].copy()
column = sdf["Sub Category"]
inventory_df = pd.concat([inventory_df,column], axis = 1)

total = np.sum(np.multiply(sdf["Total Qty"],sdf["Avg Inventory Value"]))


inventory_df['0-7 days'] = np.divide(inventory_df['0 - 7 Days Count'] * inventory_df['0 - 7 Days Avg'],total)
inventory_df['8-15 days'] = inventory_df['8 - 15 Days Count'] * inventory_df['8 - 15 Days Avg']/total
inventory_df['16-30 days'] = inventory_df['16 - 30 Days Count'] * inventory_df['16 - 30 Days Avg']/total
inventory_df['31-45 days'] = inventory_df['31 - 45 Days Count'] * inventory_df['31 - 45 Days Avg']/total
inventory_df['46-60 days'] = inventory_df['46 - 60 Days Count'] * inventory_df['46 - 60 Days Avg']/total
inventory_df['61-90 days'] = inventory_df['61 - 90 Days Count'] * inventory_df['61 - 90 Days Avg']/total
inventory_df['91-120 days'] = inventory_df['91 - 120 Days Count'] * inventory_df['91 - 120 Days Avg']/total
inventory_df['121-150 days'] = inventory_df['121 - 150 Days Count'] * inventory_df['121 - 150 Days Avg']/total
inventory_df['151-180 days'] = inventory_df['151 - 180 Days Count'] * inventory_df['151 - 180 Days Avg']/total
inventory_df['Older Than 180 Days'] = inventory_df['Older Than 180 Days Count'] * inventory_df['Older Than 180 Days Avg']/total

sum_of_inventory = inventory_df.groupby(['Sub Category']).agg({ '0-7 days': 'sum', '8-15 days' : 'sum', 
                                                                    '16-30 days' : 'sum', '31-45 days' : 'sum',
                                                                    '46-60 days' : 'sum', '61-90 days' : 'sum',
                                                                    '91-120 days' : 'sum', '121-150 days' : 'sum',
                                                                    '151-180 days' : 'sum', 'Older Than 180 Days' : 'sum',
                                                               
                                                                   })








                               
Inventory_Age_Based_Report_df = (sum_of_inventory*100).round(1)
Inventory_Age_Based_Report_df

Inventory_Age_Based_Report_df.groupby('Sub Category')['46-60 days'].sum()


Inventory_Age_Based_Report_df
```

```python
Pre_Owned_df = Formulated_df.groupby(['Product Age']).filter(lambda x: (x['Product Age'] == 'Pre Owned').any())
GC_Report_df = Pre_Owned_df.groupby('Class Name').agg({'Total Purchase Value': 'sum',
                                       'Total Qty':'sum',
                                       'Class Name' : 'count',
                                       'Value Within Limit' : 'sum',
                                       'Value Outside Limit': 'sum',
                                       'Margin' : 'sum'})

d1 = GC_Report_df.drop(index=GC_Report_df.iloc[0:1].index.tolist())
d2 = d1.drop(index=GC_Report_df.iloc[2:3].index.tolist())
d3 = d2.drop(index=GC_Report_df.iloc[9:10].index.tolist())
GC_Report_df = d3

c_report = cdf.groupby(['Class Name']).agg({'Category':'count',
                                'Total Sales':'sum'})
d11 = c_report.drop(index=c_report.iloc[3:4].index.tolist())
d12 = d11.drop(index=c_report.iloc[4:5].index.tolist())
c_report=d12


gtotal = c_report['Total Sales'].sum()

c_report['Total Sales'] = (c_report['Total Sales']/gtotal*100).round(1)
c_report = c_report.rename(index={'All Time Hits': 'All time Hits'})

GC_Report_df['Revn Benchmark'] = c_report['Total Sales']
new_index =['Volatile Games','All time Hits','Moderate','Low','NS Games','NS Junk','Junk A','Junk B']
GC_Report_df = GC_Report_df.reindex(new_index)
GC_Report_df['Stock Holdings'] = (GC_Report_df['Total Purchase Value'] / GC_Report_df['Total Purchase Value'].sum()*100).round(1)
GC_Report_df['Total Games'] = c_report['Category']
GC_Report_df['In Stock Ratio'] = (GC_Report_df['Class Name']/GC_Report_df['Total Games']*100).round(1)
GC_Report_df['Value Within Limit'] = (GC_Report_df['Value Within Limit']/GC_Report_df['Total Purchase Value']*100).round(1)
GC_Report_df['Value Outside Limit'] = (GC_Report_df['Value Outside Limit']/GC_Report_df['Total Purchase Value']*100).round(1)
GC_Report_df['Margin'] = (GC_Report_df['Margin']/GC_Report_df['Total Purchase Value']*100).round(1)

GC_Report_df.columns = ['Total Value','Total Qty','Games In Stock','% Inside Limit','% Outside Limit','Exp. Margin %', 'Revn Benchmark','Stock Holdings','Total Games','In Stock Ratio']



GC_Report_df
```
